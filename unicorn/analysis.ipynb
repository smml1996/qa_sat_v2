# Comparing Qubos Structure

```python
folders = ['sat_ratio', 'not_experiments/qubos', 'and_experiments/qubos', 'mul_experiments/qubos', '40bit_and_experiments/qubos']

from os.path import isfile, join
from os import listdir
from unicorn_utils import *

df_qubos = pd.DataFrame()
for folder in folders:
    for file in listdir(folder):
        path = f"{folder}/{file}"
        if isfile(path) and "unicorn" in file:
            stats = {
                "folder": folder.split("/")[0],
                "file": file.split(".")[0]
            }
            stats = stats | get_bqm_statistics_from_path(path)
            df_qubos = df_qubos.append(stats, ignore_index=True)
df_qubos
```

# Comparing CNFs Structure

```python
folders = ['sat_ratio', 'not_experiments/cnfs', 'and_experiments/cnfs', 'mul_experiments/cnfs', '40bit_and_experiments/cnfs']

from os.path import isfile, join
from os import listdir
from unicorn_utils import *

df_cnfs= pd.DataFrame()
for folder in folders:
    for file in listdir(folder):
        path = f"{folder}/{file}"
        if isfile(path) and "cnf" in file:
            stats = {
                "folder": folder.split("/")[0],
                "file": file.split(".")[0]
            }
            stats = stats | get_cnf_statistics_from_path(path)
            df_cnfs = df_cnfs.append(stats, ignore_index=True)
df_cnfs
```
